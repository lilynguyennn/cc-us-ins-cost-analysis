# U.S. Medical Insurance Costs

Project Goal 1: Analyze insurance costs by region

```python
#Import data and set up
import csv
with open('insurance.csv') as insurance_file:
    insurance_data = list(csv.DictReader(insurance_file))
    #print(insurance_data)
        
```

```python
#Find average insurance cost by region

southwest_total_cost = 0
southwest_num_records = 0

northeast_total_cost = 0
northeast_num_records = 0

northwest_total_cost = 0
northwest_num_records = 0

southeast_total_cost = 0
southeast_num_records = 0

for record in insurance_data:
    if record['region'] == 'southwest':
        southwest_total_cost += float(record['charges'])
        southwest_num_records += 1
    elif record['region'] == 'northeast':
        northeast_total_cost += float(record['charges'])
        northeast_num_records += 1
    elif record['region'] == 'northwest':
        northwest_total_cost += float(record['charges'])
        northwest_num_records += 1
    else:
        southeast_total_cost += float(record['charges'])
        southeast_num_records += 1
#print(southwest_total_cost, southwest_num_records)
def find_average_cost(total_cost, num_records):
    average_cost = total_cost / num_records
    return average_cost

#Southwest average
southwest_average_cost = find_average_cost(southwest_total_cost, southwest_num_records)
print('The average insurance cost for patients from the Southwest region is $' + str(southwest_average_cost))

#Northeast average
northeast_average_cost = find_average_cost(northeast_total_cost, northeast_num_records)
print('The average insurance cost for patients from the Northeast region is $' + str(northeast_average_cost))

#Northwest average
northwest_average_cost = find_average_cost(northwest_total_cost, northwest_num_records)
print('The average insurance cost for patients from the Northwest region is $' + str(northwest_average_cost))

#Southeast average
southeast_average_cost = find_average_cost(southeast_total_cost, southeast_num_records)
print('The average insurance cost for patients from the Southeast region is $' + str(southeast_average_cost))
```

Patients from the the Southeast region have the highest average cost of insurance, while patients from the Southwest region have the lowest average cost of insurance. We notice that patients from Eastern regions have higher average insurance costs than those of patients from Western regions. It is of interest to us to analyze the distribution of age and bmi in those two areas to examine if there's a noticable difference.

```python
#Age distribution in Eastern v. Western regions
#Grouping age data in lists by regions

eastern_age = []
western_age = []

for record in insurance_data:
    if 'east' in record['region']:
        eastern_age.append(record['age'])
    elif 'west' in record['region']:
        western_age.append(record['age'])
        
#print(eastern_age)
#print(western_age)
```

```python
#Counting age frequency in each region

def count_age(list):
    hist = {}
    for i in list:
        hist[i] = hist.get(i, 0) + 1
    sorted_lists = sorted(hist.items())
    x, y = zip(*sorted_lists)
    return x, y

ex, ey = count_age(eastern_age)
wx, wy = count_age(western_age)
```

```python
# Plotting age distribution in Eastern regions

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(20, 10), dpi=90)

plt.bar(ex, ey, 1.0, color='r')
plt.show()
```

```python
# Plotting age distribution in Eastern regions

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(20, 10), dpi=90)

plt.bar(wx, wy, 1.0, color='b')
plt.show()
```

```python
#Find average BMI by region

eastern_total_bmi = 0
eastern_num_recs = 0

western_total_bmi = 0
western_num_recs = 0

for record in insurance_data:
    if 'east' in record['region']:
        eastern_total_bmi += float(record['bmi'])
        eastern_num_recs += 1
    elif 'west' in record['region']:
        western_total_bmi += float(record['bmi'])
        western_num_recs += 1

def find_average_bmi(total_bmi, num_recs):
    average_bmi = total_bmi / num_recs
    return average_bmi
```

```python
#Eastern Avg BMI
print(find_average_bmi(eastern_total_bmi, eastern_num_recs))
```

```python
#Western Avg BMI
print(find_average_bmi(western_total_bmi, western_num_recs))
```

From our analysis, Eastern and Western region have similar age distribution among patients. However, the average BMI in Eastern patients is higher than that of Western patients. This may be a contributor to the higher average insurance cost in Eastern regions, as a higher BMI index may increase insurance charges.
